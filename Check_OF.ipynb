```python
import numpy as np
import pandas as pd
```

```python
df_O=pd.read_excel('Сер.xlsx', sheet_name = 'Общие формы')
df_O
```

```python
Dict_O = df_O[:1].reset_index(drop = True).to_dict('list')
Dict_O
```

```python
df_O.drop (index=df_O.index [0], axis= 0 , inplace= True)
```

```python
listIN=list(df_O.columns)
```

```python
df_Q = pd.DataFrame(columns=['Скрининговый_номер', 'Название_ОФ', 'Идентификатор', 'Текст_квери'])
df_Q
```

```python
def Data_Query(query,name,ID,text):
    new_rows = pd.DataFrame({'Скрининговый_номер': query['SUBJ_SCREEN'].tolist(),
'Название_ОФ': query[name].to_list(),
 'Идентификатор': [ID for i in range(len(query['SUBJ_SCREEN'].tolist()))],
 'Текст_квери': [text for i in range(len(query['SUBJ_SCREEN'].tolist()))]})
    return new_rows
```

```python
df_Ch = pd.DataFrame(index=range(0, df_O.shape[0]+1))
df_Ch
```

```python
def Data_Check(query,ID,data):
    data[ID] = 'Проверено'
    data.loc[:,ID].iloc[query.index.tolist(),] = 'Требует уточнения специалиста по обработке данных'
```

```python
df_M = pd.DataFrame(columns=['Скрининговый_номер', 'Название_ОФ', 'Идентификатор', 'Заголовок_в_эИРК','Данные_в_поле','Возможная_ошибка'])
df_M
```

```python
def Data_Management(query,name,ID,text):
    new_rows_M = pd.DataFrame({'Скрининговый_номер': query['SUBJ_SCREEN'].tolist(),
'Название_ОФ': query[name].to_list(),
 'Идентификатор': [ID for i in range(len(query['SUBJ_SCREEN'].tolist()))],
 'Заголовок_в_эИРК': [Dict_O.get(ID) for i in range(len(query['SUBJ_SCREEN'].tolist()))],
 'Данные_в_поле': query[ID].to_list(),
 'Возможная_ошибка': [text for i in range(len(query['SUBJ_SCREEN'].tolist()))]})
    return new_rows_M
```

Проверка НЯ (исход и дата окончания). Исход должен быть известен, если указана дата окончания:

```python
query = df_O.loc[(df_O['03_AEENDTC']==df_O['03_AEENDTC'])&(df_O['03_AEOUT'] == 'Неизвестно')]
query
```

```python
df_Q = pd.concat([df_Q, Data_Query(query,'03_AETERM', '03_AEOUT', text = 'Пожалуйста, уточните исход НЯ')], ignore_index=True)

df_Ch = pd.concat([df_Ch, Data_Check(query, '03_AEOUT',df_Ch)], ignore_index=True)

df_M = pd.concat([df_M, Data_Management(query, '03_AETERM', '03_AEOUT', text = 'Указан некорректный исход НЯ (Неизвестен исход, но дата окончания существует)')], ignore_index=True)
```
