Ввести клинические проектные клинические оценки:

```python
AE = 'Отклонение, связано с НЯ'
MH = 'Отклонение, связано с анамнезом'
```

```python
df_Q = pd.DataFrame(columns=['Скрининговый_номер','ID_поля', 'Заголовок_в_эИРК', 'Отклонение', 'Прикреплённая_ОФ','Существует_ОФ'])
df_Q
```

Функция заполнения датасета:

```python
def Data_OF(query,ID,lis1,lis2):
    a=query.index - 1
    b = (df_V.iloc[a.tolist(), df_V.columns.get_loc(ID)+1] + " " + query['SUBJ_SCREEN'].astype(str)).to_list()
    Availability = []
    for i in range(len(b)):
        if b[i] in lis1 or b[i] in lis2:
            Availability.append('TRUE')
        else:
            Availability.append('FALSE')
    new_rows = pd.DataFrame({'Скрининговый_номер': query['SUBJ_SCREEN'].tolist(),
 'ID_поля': [ID for i in range(len(query['SUBJ_SCREEN'].tolist()))],
         'Заголовок_в_эИРК': [Dict_V.get(df_V.columns[df_V.columns.get_loc(ID) - 2]) for i in range(len(query['SUBJ_SCREEN'].tolist()))],
        'Отклонение': query[ID].to_list(),
       'Прикреплённая_ОФ': df_V.iloc[a.tolist(), df_V.columns.get_loc(ID)+1].tolist(),
        'Существует_ОФ': Availability})
    Availability = []
    return new_rows
```
